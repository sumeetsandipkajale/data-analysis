```python
# df['Last_Name'] = df['Last_Name'].str.lstrip("...")
# df['Last_Name'] = df['Last_Name'].str.lstrip("/")
# df['Last_Name'] = df['Last_Name'].str.rstrip("_")
df['Last_Name'] = df['Last_Name'].str.strip("123./_")
df
```

```python
# to remove the unnecesary characters in between
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA_Z0-9]','')

# to convert all elements of column in string format
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df.dtypes

# to make phone number more good to see
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' +x[6:10])
df

# to remove null values
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('a--','')
df

```

```python
# split address in seperate columns
df[["street Address" , "state" , "zipCode"]] =  df['Address'].str.split(',',2 , expand = True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')

df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')

df
```

```python
# df = df.replace("N/a" , "")
# df = df.replace("NaN" , "")

# to drop all null with ''
df = df.fillna('')

```

```python
df
```

```python
only call the people who said N in the column Do_Not_Contact

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes' , 'Y')


df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No' , 'N')

df
# drop who said yes

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x,inplace = True)
        
# drop who has said nothing

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == '':
        df.drop(x,inplace = True)

df
```

```python
# drop null of phone numbers

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x,inplace = True)

# we can also drop in this way
# df = df.dropna(subset = 'Phone_Number') , inplace = True
```
